### Blobs

```python
def f(x):
    # y = mx + b
    return 0.3*x + 0.2


class Point:
    def __init__(self, width, height):
        self.x = np.random.uniform(-1, 1, width)
        self.y = np.random.uniform(-1, 1, height)
        self.bias = 1
        self.labels = []
    
    def point(self):
        for x, y in zip(self.x, self.y):
            lineY = f(x)
            if y > lineY:
                self.labels.append(1)
            else:
                self.labels.append(-1)

        return self.labels


points = Point(500, 500)
labels = points.point()
x = points.x
y = points.y
inputs = pd.DataFrame({'x':x, 'y':y})
```
